# Univariate EDA for - 
-  time 
- amount

```python
# univariate EDA for amount 

plt.figure(figsize=(10,5))
sns.histplot(df['Amount'], bins=50, kde=True)
plt.title("Distribution of Transaction Amounts")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(10,5))
sns.histplot(df['Time'], bins = 50, kde=True)
plt.title("Distribution of Transaction Time")
plt.xlabel("Time (seconds)")
plt.ylabel("Frequency")
plt.show()
```

# Interpretations - 

# Amount column
1. The Amount feature is highly right-skewed, with the majority of transactions having very small values and a long tail of large outliers.
2. Needs log transformation
3. Extreme outliers exist

# Time column
1. The dataset spans ~2 days
2. Multimodal distribution
3. Distinct peaks at different time ranges
4. Not normally distributed
5. Time by itself may not be very predictive, but 
    - fraud patterns can change by time of data 
    

# Bivariate EDA for amount one with -
- Bivariate EDA
- class = 0 (no fraud)
- class  1 (fraud)

```python
plt.figure(figsize=(10,8))
sns.histplot(df[df['Class']==0]['Amount'], bins=50)
plt.title("Bivariate EDA for Amount with Class = 0 (No Fraud)")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()


plt.figure(figsize=(10,8))
sns.histplot(df[df['Class']==1]['Amount'], bins=50, color='red')
plt.title("Bivariate EDA for Amount with Class = 1 (Fraud)")
plt.xlabel("Amount")
plt.ylabel("Frequency") 
plt.show()
```

# Interpretations - 

- Bivariate EDA for Amount with Class = 0 (No Fraud)
1. A huge spike near Amount â‰ˆ 0
2. A very long right tail up to ~25,000, outliers exist meaning a few large transactions exists
3. Most transactions are small everyday amounts. Non-fraud transactions are small payments.
    Hence most values cluster at low amounts 

- Bivariate EDA for Amount with Class = 1 (Fraud)
1. Still many small-amount frauds
2. Fraud appears across a wider range of amounts

Fraud happens at all amounts,but the distribution shape differs.
Amount alone cannot detect fraud, needs pca componnets to detect fraud.

creating log transformed amount column

```python
import numpy as np
df['Amount_log'] = np.log1p(df['Amount'])
df.columns
```

# Bivariate EDA for amount_log with class = 0 and class = 1

```python
plt.figure(figsize=(10,5))
sns.histplot(df[df['Class']==0]['Amount_log'], bins=50)
plt.title("Bivariate EDA for Amount_log with Class = 0 (No Fraud)")
plt.xlabel("Log Transformed Amount")
plt.ylabel("Frequency")
plt.show()


plt.figure(figsize=(10,5))
sns.histplot(df[df['Class']==1]['Amount_log'], bins=50, color='red')
plt.title("Bivariate EDA for Amount_log with Class = 1 (Fraud)")
plt.xlabel("Log Transformed Amount")
plt.ylabel("Frequency")
plt.show()
```

# inference - 

- Log Amount â€“ Class = 0
1. Distribution is now spread out 
2. Most normal distributions cluster heavily around low to mid log amounts.

- Log Amount â€“ Class = 1 (Fraud)
1. Fraud transactions appear across multiple log-amount ranges.
2. No single â€œsafe amount zoneâ€
3. Fraud overlaps with normal behavior.

- Transaction Amount is highly skewed and requires log transformation.
- Fraud transactions occur across both small and medium amounts, not only large ones.
- There is significant overlap between fraud and non-fraud distributions.
- Amount alone is insufficient to detect fraud. Fraud detection will require multivariate patterns, not single-feature rules.
